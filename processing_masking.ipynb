```python
import numpy as np
import cv2
import json
```

```python
f = open('dataset/Iris and Needle Segmentation.v14-layers-sorted.coco-segmentation/valid/_annotations.coco.json', 'r')
data = json.load(f)
```

```python
img_dir = 'dataset/Iris and Needle Segmentation.v14-layers-sorted.coco-segmentation/valid/'
mask_dir = 'dataset/Iris and Needle Segmentation.v14-layers-sorted.coco-segmentation/valid_mask/'
```

```python
images = data['images']
annotations = data['annotations']
```

```python
for image in images:
    mask = np.zeros((image['height'], image['width']))
    for annotation in annotations:
        if annotation['image_id'] == image['id']:
            seg = annotation['segmentation']
            seg = np.array(seg).reshape((-1, 1, 2)).astype(np.int32)
            cv2.fillPoly(mask, [seg], 255)
    cv2.imwrite(mask_dir + image['file_name'], mask)
    print(image['file_name'])
```
